## 1. Find the 10 countries with most projects


```python
import numpy as np
import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
#
# Create the world bank dataframe
#
df = pd.read_json('data/world_bank_projects.json')
```

```python
#
# Print the head.
#
df.head()
```

```python
#
# Obtain a list of the 10 countries with the most projects.
#
df.groupby(['countryname']).size().sort_values(ascending=False).head(10)
```

## 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
#
# Load the JSON.
#
json_str = json.load((open('data/world_bank_projects.json')))

#
# Flatten the string into a dataframe on the 'mjtheme_namecode' field.
#
theme_df = json_normalize(json_str, 'mjtheme_namecode', ['id'])

#
# Print the head.
#
theme_df.head()
```

```python
#
# Obtain the 10 countries with the most projects.
#
theme_df.name.value_counts().sort_values(ascending=False).head(10)
```

** As you can see above, 122 projects do not have a name string associated with it.**

## 3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
#
# Print the head.
#
theme_df.head()
```

```python
#
# Drop the ID column.
#
theme_df = theme_df.drop('id', axis=1)

#
# Print the head.
#
theme_df.head()
```

```python
#
# Get unique combination of code and name from dataset without empty strings.
#
unique_df = theme_df[theme_df.name != ''].drop_duplicates()

#
# Take the codes from the original df and merge with the unique df on the 'code' column.
#
merged_df = pd.DataFrame(theme_df.code).merge(unique_df, on='code', how='left')

#
# Perform a recount of the 10 countries with the most projects.
#
merged_df.name.value_counts().sort_values(ascending=False).head(10)
```

** As you can see, the empty names have been replaced and the counts have been updated to reflect the new totals. **
